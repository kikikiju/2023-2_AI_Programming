```python
import pandas as pd

df1 = pd.read_csv('data_1.csv', encoding='cp949')
df2 = pd.read_csv('data_2.csv', encoding='cp949')
df3 = pd.read_csv('data_3.csv', encoding='cp949')
```

```python
df1_a = df1[df1['감정'] == 'angry']
df1_b = df1[df1['감정'] == 'sadness']
df1_c = df1[df1['감정'] == 'fear']
df1_d = df1[df1['감정'] == 'surprise']
df1_e = df1[df1['감정'] == 'happiness']

DF_1 = pd.concat([df1_a,df1_b,df1_c,df1_d,df1_e])

DF_1['감정'] = DF_1['감정'].replace('angry', '분노')
DF_1['감정'] = DF_1['감정'].replace('sadness', '슬픔')
DF_1['감정'] = DF_1['감정'].replace('fear', '불안')
DF_1['감정'] = DF_1['감정'].replace('surprise', '불안')
DF_1['감정'] = DF_1['감정'].replace('happiness', '행복')

DF_1
```

```python
df2_a = df2[df2['감정'] == '분노']
df2_b = df2[df2['감정'] == '슬픔']
df2_c = df2[df2['감정'] == '불안']
df2_d = df2[df2['감정'] == '기쁨']

DF_2 = pd.concat([df2_a,df2_b,df2_c,df2_d])
DF_2.reset_index(drop=True, inplace=True)
DF_2['감정'] = DF_2['감정'].replace('기쁨', '행복')
DF_2.head()
```

```python
df3_a = df3[df3['감정'] == '분노']
df3_b = df3[df3['감정'] == '슬픔']
df3_c = df3[df3['감정'] == '불안']
df3_d = df3[df3['감정'] == '기쁨']

DF_3 = pd.concat([df3_a,df3_b,df3_c,df3_d])
DF_3.reset_index(drop=True, inplace=True)
DF_3['감정'] = DF_3['감정'].replace('기쁨', '행복')
DF_3.head()
```

```python
DF = pd.concat([DF_1,DF_2,DF_3])
DF.head()
```

```python
DF['감정'].value_counts()
```

```python
DF_copy = DF.copy()

# 분노 -> 0, 슬픔 -> 1, 불안 -> 2, 행복 -> 3
DF_copy.loc[(DF_copy['감정'] == "분노"), '감정'] = 0

DF_copy.loc[(DF_copy['감정'] == "슬픔"), '감정'] = 1

DF_copy.loc[(DF_copy['감정'] == "불안"), '감정'] = 2

DF_copy.loc[(DF_copy['감정'] == "행복"), '감정'] = 3

DF.loc[:, '감정'] = DF_copy['감정']
DF.reset_index(drop=True, inplace=True)

# 데이터 확인
DF.head()
```

```python
DF.to_csv('new_dataset.csv', index=False)
```
