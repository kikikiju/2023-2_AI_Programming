```python
import pandas as pd

df_1 = pd.read_csv('data/data-preprocess/data-1.csv', encoding='cp949')
df_2 = pd.read_csv('data/data-preprocess/data-2.csv', encoding='cp949')
df_3 = pd.read_csv('data/data-preprocess/data-3.csv', encoding='cp949')
df_4 = pd.read_csv('data/data-preprocess/data-4.csv', encoding='cp949')
df_5 = pd.read_csv('data/data-preprocess/data-5.csv', encoding='cp949')
df_6 = pd.read_csv('data/data-preprocess/data-6.csv', encoding='cp949')
df_7 = pd.read_csv('data/data-preprocess/data-7.csv', encoding='cp949')
```

```python
# 3개로 나눠져있는 데이터 합쳐주기
df_a = pd.concat([df_1,df_2,df_3])

# 대소문자 구분 없애기
df_a.loc[:, '감정'] = df_a['감정'].str.lower()

# 중립 감정은 제거해주기
df_a = df_a[df_a['감정'] != 'neutral']

df_a_copy = df_a.copy()

# 분노 -> 0, 슬픔 -> 1, 불안 -> 2, 행복 -> 3
df_a_copy.loc[(df_a_copy['감정'] == "angry"), '감정'] = 0
df_a_copy.loc[(df_a_copy['감정'] == "disgust"), '감정'] = 0
df_a_copy.loc[(df_a_copy['감정'] == "sadness"), '감정'] = 1
df_a_copy.loc[(df_a_copy['감정'] == "fear"), '감정'] = 2
df_a_copy.loc[(df_a_copy['감정'] == "surprise"), '감정'] = 2
df_a_copy.loc[(df_a_copy['감정'] == "happiness"), '감정'] = 3

df_a.loc[:, '감정'] = df_a_copy['감정']
df_a.reset_index(drop=True, inplace=True)

# 데이터 확인
df_a
```

```python
df_b = pd.concat([df_4,df_5,df_6,df_7])

# 중립 감정은 제거해주기
df_b = df_b[df_b['감정'] != '중립']

df_b_copy = df_b.copy()

# 분노 -> 0, 슬픔 -> 1, 불안 -> 2, 행복 -> 3
df_b_copy.loc[(df_b_copy['감정'] == "분노"), '감정'] = 0
df_b_copy.loc[(df_b_copy['감정'] == "혐오"), '감정'] = 0

df_b_copy.loc[(df_b_copy['감정'] == "슬픔"), '감정'] = 1
df_b_copy.loc[(df_b_copy['감정'] == "상처"), '감정'] = 1

df_b_copy.loc[(df_b_copy['감정'] == "공포"), '감정'] = 2
df_b_copy.loc[(df_b_copy['감정'] == "놀람"), '감정'] = 2
df_b_copy.loc[(df_b_copy['감정'] == "불안"), '감정'] = 2
df_b_copy.loc[(df_b_copy['감정'] == "당황"), '감정'] = 2

df_b_copy.loc[(df_b_copy['감정'] == "행복"), '감정'] = 3
df_b_copy.loc[(df_b_copy['감정'] == "기쁨"), '감정'] = 3

df_b.loc[:, '감정'] = df_b_copy['감정']
df_b.reset_index(drop=True, inplace=True)

# 데이터 확인
df_b
```

```python
# 전체 데이터 합치기
df = pd.concat([df_a,df_b])
df
```

```python
import numpy as np

# 가장 적은 클래스의 개수 구하기
class_counts = df['감정'].value_counts()
min_class_count = class_counts.min()

# 랜덤 시드 고정
np.random.seed(42)

# 각 클래스에서 랜덤으로 샘플 추출
sampled_data = []
for emotion_class in class_counts.index:
    class_data = df[df['감정'] == emotion_class]
    sampled_data.append(class_data.sample(min_class_count))

# 새로운 데이터프레임 생성
balanced_df = pd.concat(sampled_data)
balanced_df.reset_index(drop=True, inplace=True)

# 결과 확인
print(balanced_df)
```

```python
class_counts = balanced_df['감정'].value_counts()
class_counts
```

```python
balanced_df.to_csv("data_main.csv", index=False)
```
